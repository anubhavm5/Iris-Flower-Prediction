```python
df=pd.DataFrame(i.data,columns=i.feature_names)
df["species"]=i.target
```

```python
setosa=df[df["species"]==0]
versi=df[df["species"]==1]
virgi=df[df["species"]==2]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.xlabel("Length Of Sepal")
plt.ylabel("Width of Sepal")
plt.scatter(setosa["sepal length (cm)"],setosa["sepal width (cm)"],color="red", label="setosa")
plt.scatter(versi["sepal length (cm)"],versi["sepal width (cm)"],color="blue", label="versi")
plt.scatter(vergi["sepal length (cm)"],vergi["sepal width (cm)"],color="green", label="vergi")
plt.legend()
```

```python
plt.xlabel("Length Of Petal")
plt.ylabel("Width of Petal")
plt.scatter(setosa["petal length (cm)"],setosa["petal width (cm)"],color="red", label="setosa")
plt.scatter(versi["petal length (cm)"],versi["petal width (cm)"],color="blue", label="versi")
plt.scatter(vergi["petal length (cm)"],vergi["petal width (cm)"],color="green", label="vergi")
plt.legend()
```

```python
x=df.drop(["species"],axis=1)
```

```python
y=df.species
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
len(train_x),len(test_x),len(train_y),len(test_y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
import warnings
warnings.filterwarnings("ignore")
```

```python
model=KNeighborsClassifier()
```

```python
model.fit(train_x,train_y)
```

```python
pred=model.predict([[5.8,5.6,1.4,6.7]])
if pred==0:
    print("setosa")
elif pred==1:
    print("versicolor")
else:
    print("virginica")
```

```python
model.score(test_x,test_y)
```

```python
model.predict(test_x)==test_y
```
